```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/blidd-google/cf3-preferrest-experiments/main/v1-logs.csv')
dfv2 = pd.read_csv('https://raw.githubusercontent.com/blidd-google/cf3-preferrest-experiments/main/v2-logs.csv')
```

# 1st gen cold start performance data


```python
# remove superfluous log entries
df = df.loc[df['severity'] == 'INFO']

data_simple = df.loc[(df['jsonPayload.name'] == 'simple')]['jsonPayload.elapsedMs']
data_off = df.loc[(df['jsonPayload.name'] == 'preferrestoffv1')]['jsonPayload.elapsedMs']
data_on = df.loc[(df['jsonPayload.name'] == 'preferrestonv1')]['jsonPayload.elapsedMs']
```

```python
plt.title(f'Cold start performance (simple v1 HTTP function), n={data_simple.size}')
plt.xlabel('Cold start execution time (ms)')
plt.ylabel('Frequency')

data_simple.hist(bins=30, alpha=0.5)
```

```python
plt.title(f'Cold start performance (preferRest disabled v1 function), n={data_off.size}')
plt.xlabel('Cold start execution time (ms)')
plt.ylabel('Frequency')

data_off.hist(bins=range(0, 12000, 200), alpha=0.5)
```

```python
plt.title(f'Cold Start Performance (preferRest enabled v1 function), n={data_on.size}')
plt.xlabel('Cold start execution time (ms)')
plt.ylabel('Frequency')

data_on.hist(bins=range(0, 12000, 200), alpha=0.5)
```

```python
plt.hist(data_on, label=f'preferRest enabled (n={data_on.size})', bins=range(0, 12000, 200), alpha=0.5);
plt.hist(data_off, label=f'preferRest disabled (n={data_off.size})', bins=range(0, 12000, 200), alpha=0.5);
plt.legend(loc='upper right');

plt.title('Cold start execution time frequencies (v1 comparison)')
plt.xlabel('Cold start execution time (ms)')
plt.ylabel('Frequency')
```
